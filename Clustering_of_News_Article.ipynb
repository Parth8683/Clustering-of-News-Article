```python
#Importing All the libraries for the process
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import adjusted_mutual_info_score
%matplotlib inline
```

```python
#Applying random seed so that everytime the same of examples are taken
np.random.seed(1771)
```

```python
#Question 1 reading the CSV files as it is.
truelabels = pd.read_csv("bbcsport_classes.csv",header= None).values
terms = pd.read_csv("bbcsport_terms.csv",header=None).values
X = pd.read_csv("bbcsport_mtx.csv",header = None).values
```

```python
#Question 2
#Using K means Cluster with five clusters as in K mean cluster Eucledian distance is inbuild. 5 clusters have been used here. 
model=KMeans(n_clusters=5)
model.fit(X)
```

```python
#Putting the lables and centroid_1
labels=model.predict(X)
centroid_1=model.cluster_centers_
```

```python
#Two matrices has been created for Adjusted rand Index and Adjusted Mututal Information
Truelabels1=np.squeeze(np.asarray(truelabels))

ARI_list = []
AMI_list = []
```

```python
#K means clusters has been used with 50 iteration and 5 clusters and then compare it with truelabels. Before comparing truelabes 
# needed to turn into array for comparison. After Adjusted rand score and Adjusted mutual information has been used. As 50 
#iteration has been used. For each iteration one Adjusted rand score and one Adjusted mutual information has been provided.
for i in range (50):
    model=KMeans(n_clusters=5)
    model.fit(X)

    labels=model.predict(X)
    centroid_1=model.cluster_centers_


    Truelabels1=np.squeeze(np.asarray(truelabels))


    ARI=adjusted_rand_score(Truelabels1, labels)
    AMI=adjusted_mutual_info_score(Truelabels1,labels)
    ARI_list = np.append(ARI_list, ARI)
    AMI_list = np.append(AMI_list, AMI)
    
print(ARI_list)
print(AMI_list)
```

```python
# Averaging has been done for better interpretation
np.average(ARI_list)
# Adjusted rand index shows is the clusters matches with truelabel or not. It's stays between 0 to 1. The more the result the 
#better the model 0.09 of Adjusted rand index shows it's a not a good model. The model can't learn the points and cluster 
#them properly. It's just putting all the custers on random basis.
```

```python
# Averaging has been done for better interpretation of the model
np.average(AMI_list)
#Adjusted Mutual information measures the agreement between truelabels and clusters. The higher the better model. .17 of adjusted
#mutual information is slightly higher compared to adjusted rand index. But, still it's near 0. So, it also evaluates the model
#as poor performing
```

```python
# import the necessary modules here Part B
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn import metrics

print(X.shape)
```

```python
#normalize our data
from sklearn.preprocessing import scale
Xnorm = scale(X)
```

```python
#perform PCA using sklearn PCA implementation

from sklearn.decomposition import PCA
pca = PCA()
pca.fit(Xnorm)

#The amount of variance that each PC explains
var= pca.explained_variance_ratio_
print(var)

#Cumulative Variance explains
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
print(var1)
plt.plot(var1)
plt.xlabel("Principal components")
plt.ylabel("Variance captured")
```

```python
# The dimension that captures 95% of the varience 
index1=np.where(var1>=95)
print(index1)
# Minimumm 554 dimensions will be needed from the data for capturing the varience of 95%
```

```python
# The dimension that captures 98% of the varience
index2=np.where(var1>=98)
print(index2)
#Minimum ^626 dimension needed for capturing 98% of the varience.
```
